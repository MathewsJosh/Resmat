Deseja-se remover o prego da madeira aplicando-se uma força no eixo horizontal. Um obstáculo em A impede o acesso direto, de tal modo que duas forças, uma F de 400lb e outra P, são aplicadas por cabos, como mostrado na figura em anexo (atividade 01.png).

Primeira etapa (resolução manual)
Calcule a magnitude de P necessária para garantir que a resultante T seja direcionada no eixo horizontal. Encontre também a intensidade T.


Segunda etapa (resolução computacional)
Investigue o efeito da distância entre o prego e o obstáculo nas duas forças P e T. Substitua a distância de 8'' por uma variável d. Plote os gráficos de P e T como funções de d, dentro do intervalo 2''<d<12''.


atividade 01.png

```python
import numpy as np
import math as mpl
import matplotlib.pyplot as plt
import pylab as pl
```

```python
# Função que calcula os valores de P e T e os salva em um Array para depois plota-los
forca_vals = []
resultante_vals = []

def calcula_pt():
  for i in range(3, 12):
    # Calculando o angulo em 1 e 2
    angulo1_rad = np.arctan(4/i)
    angulo2_rad = np.arctan(6/i)

    # Calculando cossenoX e senoX para 1 e para 2
    cosseno1 = np.cos(angulo1_rad)
    seno1 = np.sin(angulo1_rad)
    cosseno2 = np.cos(angulo2_rad)
    seno2 = np.sin(angulo2_rad)

    # Calculando FX e FY para 2
    Fx2 = 400*np.cos(angulo2_rad)
    Fy2 = 400*np.sin(angulo2_rad)

    # Primeira Equação de Equilibrio, P
    P = Fy2/seno1
    forca_vals.append(P)

    # Segunda Equação de equilibrio, T
    T = cosseno1 * P + Fx2
    resultante_vals.append(T)
```

```python
calcula_pt()
```

```python
forca_vals, resultante_vals
```

```python
def plot_results(x, label, labelY, title):
    i = range(3,12)

    plt.figure(figsize = (8, 5))
    
    plt.plot(i, x,'o-', label=label)
    plt.xlim(3,11)
    plt.ylim(np.amin(x) - 30, np.amax(x) + 30)
    plt.grid(True)

    # Personalização do gráfico (label, fontes, titulos e subtitulos)
    plt.title(title)
    plt.xlabel("Distância", labelpad = 18, fontsize = 18)
    plt.xticks(fontsize = 12)
    plt.ylabel(labelY, labelpad = 18, fontsize = 18)
    plt.yticks(fontsize = 12)
    plt.legend(loc = "best", fontsize = 12, frameon = True)

    #plt.show()
```

```python
def plot_results_2in1(title):
    i = range(3,12)
    #forca_vals, resultante_vals
    plt.figure(figsize = (8, 5))
    
    plt.plot(i, forca_vals,'o-', label="Magnitude P")
    plt.plot(i, resultante_vals,'o-', label="Intensidade T")
    plt.xlim(3,11)
    plt.ylim(400, 900)
    plt.grid(True)

    # Personalização do gráfico (label, fontes, titulos e subtitulos)
    plt.title(title,fontsize= 14)
    plt.xlabel("Distância", labelpad = 18, fontsize = 18)
    plt.xticks(fontsize = 12)
    plt.ylabel("Unidade de força", labelpad = 18, fontsize = 18)
    plt.yticks(fontsize = 12)
    plt.legend(loc = "best", fontsize = 12, frameon = True)

    #plt.show()
```

```python
plot_results_2in1("Gráfico de magnitude e intensidade")
```

```python
plot_results(forca_vals, "Magnitude P", "Força P", "Gráfico da Magnitude P")
plot_results(resultante_vals, "Intensidade T", "Intensidade T", "Gráfico da Intensidade T")
```
